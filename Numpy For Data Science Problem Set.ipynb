```python
# Exercise 1 — Replace Loop with NumPy

"""
Problem:
    nums = [1, 2, 3, 4, 5]

Multiply every value by 2

Step 1 (Python way):
    [x * 2 for x in nums]

Your Task:
    Do this using NumPy
"""

import numpy as np

nums = [1, 2, 3, 4, 5]
arr = np.array(nums)
numsx2 = arr*2
print(numsx2)
```

```python
# Exercise 2 — Filtering

"""
Problem:
    nums = [10, 20, 30, 5, 3]

    Return only values greater than 10

Constraint:
    No loops
    Must use NumPy
"""

import numpy as np
nums = [10, 20, 30, 5, 3]
arr = np.array(nums)
filtered = arr > 10
print(arr[filtered])
```

```python
# Exercise 3 — Basic Stats

"""
Problem:
    nums = [10, 20, 30, 40, 50]

Compute:
    mean
    sum
    standard deviation
"""

import numpy as np

nums = [10, 20, 30, 40, 50]
arr = np.array(nums)

summary = {
    "mean": arr.mean(),
    "sum": arr.sum(),
    "std": arr.std()
}

print(summary)
```

```python
# Exercise 4 — 2D Arrays

"""
Problem:
    data = [
        [25, 50000],
        [30, 60000],
        [22, 45000]
    ]

Convert to NumPy array and:

    Get all ages
    Get all salaries
    Compute average salary
"""

import numpy as np

data = [
        [25, 50000],
        [30, 60000],
        [22, 45000]
    ]

data_arr = np.array(data)

ages = data_arr[:,0]
salaries = data_arr[:, 1]
avg_salaries = salaries.mean()

summary = {
    "ages": ages,
    "salaries": salaries,
    "avg_salary": avg_salaries
}

print(summary)
```

```python
# Exercise 5 — Vectorized Logic

"""
Problem:
    nums = [5, 15, 25, 2, 8]

Create a new array:
    If value > 10 → label "high"
    Else → label "low"

Constraints: No loops
"""
import numpy as np 

nums = [5, 15, 25, 2, 8]
npnums = np.array(nums)

labels = np.where(npnums > 10, "high", "low")
print(labels)

# Bonus Exercise — Multi-Condition Filtering

"""
Problem:
    nums = [5, 12, 18, 3, 25, 7]

    Return values that are:
        greater than 10
        AND less than 20

    Constraints:
        Use NumPy
        No loops
"""
nums = [5, 12, 18, 3, 25, 7]
npnums = np.array(nums)

values = npnums[(npnums > 10) & (npnums < 20)]
print(values)

# Bonus Exercise — Multi-Condition Labeling

"""
Problem:
    nums = [5, 12, 18, 3, 25, 7]

    Create labels:
        "low" → <= 10
        "medium" → > 10 and <= 20
        "high" → > 20

    Constraints:
        Use Numpy
        No loops
"""
nums = [5, 12, 18, 3, 25, 7]
npnums = np.array(nums)

multi_labels = np.select([(npnums <= 10), (npnums > 10) & (npnums <= 20), (npnums > 20)], ["low", "medium", "high"], default="unknown")
print(multi_labels)
```
