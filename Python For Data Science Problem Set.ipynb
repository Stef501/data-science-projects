```python
## Exercise 1 — Debugging + Scope
"""
This function is supposed to count how many values are above a threshold:

    def count_above(nums, threshold):
        count = 0
    
        for num in nums:
            if num > threshold:
                count += 1
    
        return count

    count = count_above([10, 20, 30, 5], 15)
    print(count)

Modify this so that:
    It also keeps track of the values above the threshold
    Returns BOTH:
        count
        list of values
"""
def count_above(nums, threshold):
    count = 0
    values = []
    
    for num in nums:
        if num > threshold:
            count += 1
            values.append(num)
   
    return {"count": count, "values": values}

count = count_above([10, 20, 30, 5], 15)
print(count)
```

```python
## Exercise 2 — Function Design
"""
Write a function:
    def summarize_data(nums):
        
That returns:
    mean
    min
    Max

Output Format:
    {
    "mean": ...,
    "min": ...,
    "max": ...
    }  
"""

def summarize_data(nums):
    if len(nums) == 0:
        return {"mean": none, "min": none, "max": none}
    
    mean_val = sum(nums)/len(nums)

    return{
        "mean": mean_val, 
        "min": min(nums), 
        "max": max(nums)
    }

print(summarize_data([1,2,3,4,5]))    
```

```python
## Exercise 3 — Lambda + Transformation
"""
You have:
    nums = [100, 200, 300, 400]

Create a new list where:
    each value is converted to thousands (divide by 1000)

Constraint:
    Must use lambda
"""

nums = [100, 200, 300, 400]
nums_k = list(map(lambda x: x / 1000, nums))

print(nums_k)

# Also done in list comprehension without lambda constraint
nums_k = [item/1000 for item in nums]
print(nums_k)
```

```python
## Exercise 4 — Realistic Data Cleaning (Python only)
"""
data = [
    {"name": "A", "age": 25},
    {"name": "B", "age": None},
    {"name": "C", "age": 30},
    {"name": "D", "age": None}
]

Write a function that:

removes entries with missing age
returns cleaned list
"""
data = [
    {"name": "A", "age": 25},
    {"name": "B", "age": None},
    {"name": "C", "age": 30},
    {"name": "D", "age": None}
]

def data_clean(dt):
    
    cleaned = []
    
    for item in dt:
        if item.get("age") is not None:
            cleaned.append(item)
        
    return cleaned

print(data_clean(data))

# Efficient Version
def data_clean(dt):
    return [item for item in dt if item.get("age") is not None]

print(data_clean(data))
```

```python
## Exercise 5 — Intermediate (Thinking Like pandas)
"""
transactions = [
    {"user": "A", "amount": 100},
    {"user": "B", "amount": 200},
    {"user": "A", "amount": 50},
    {"user": "B", "amount": 300},
    {"user": "C", "amount": 400}
]

Compute total spend per user:

Expected output:
{
    "A": 150,
    "B": 500,
    "C": 400
}
"""
transactions = [
    {"user": "A", "amount": 100},
    {"user": "B", "amount": 200},
    {"user": "A", "amount": 50},
    {"user": "B", "amount": 300},
    {"user": "C", "amount": 400}
]

def total_spend(transactions):

    result = {}
    
    for item in transactions:  
        user = item["user"]
        amount = item["amount"]

        if user not in result:
            result[user] = 0

        result[user] += amount
            
    return result
      
print(total_spend(transactions))   

# Efficient Version
def total_spend(transactions):

    result = {}
    
    for item in transactions:
        user = item["user"]
        result[user] = result.get(user,0) + item["amount"]
        
    return result
    
print(total_spend(transactions))
```
